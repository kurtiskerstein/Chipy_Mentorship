# Pulling bulk Petroleum Data in JSON format
http://api.eia.gov/bulk/PET.zip

## What information is available from the Dataset?

# JSON object image

```python
data = pd.read_json('./PET.txt',orient='columns',lines=True,convert_dates=['start','end'])
data.info()
```

```python
data.head()
```

### There are two JSON object types: Series and category

### Category Data

```python
# Figure out what data is available for the categories
category_data = data[pd.notnull(data['category_id'])]
category_data.info()
```

#### Structured with a parent category and child categories. 

```python
category_data[['name','category_id','childseries','parent_category_id']].head()
```

### Series data
- Find the structure of the series data

```python
series_data = data[pd.notnull(data['series_id'])]
series_data.info()
```

```python
series_data = (series_data[['series_id','name','units','description','unitsshort','f','start','end','last_updated']]
 .set_index('series_id')
)
series_data['start']= pd.to_datetime(series_data['start'])
series_data['end']= pd.to_datetime(series_data['end'])
series_data.head()
```

### Annual Data

```python
series_data[series_data.index.str.contains('^PET.*.A$',regex=True)]
```

### Monthly Data

```python
series_data[series_data.index.str.contains('^PET.*.M$',regex=True)]
```

### Weekly Data

```python
series_data[series_data.index.str.contains('^PET.*.W$',regex=True)]
```

### Daily Data

```python
series_data[series_data.index.str.contains('^PET.*.D$',regex=True)]
```

### Monthly Price data

```python
series_data[series_data.index.str.contains('^PET.R[WC].*.M$',regex=True)]
```

### Daily Price Data

```python
series_data[series_data.index.str.contains('^PET.R[WC].*.D$',regex=True)]
```
